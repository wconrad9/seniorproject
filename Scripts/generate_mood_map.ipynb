```python
value_progression = []

for path in imageset:
    rgb_img = imread(path)

    hsv_img = rgb2hsv(rgb_img)
    value_img = hsv_img[:,:,2]
    
    value_img = value_img*mask
    
    value_progression.append(np.mean(value_img))
    
```

```python
quarters = []
for i in range(1,5):
    boundary0 = (i-1)*len(value_progression)//4
    boundary1 = i * len(value_progression)//4
    quarters.append(value_progression[boundary0:boundary1])

for entry in quarters:
    print(np.mean(entry))
```
