```python
import librosa
import librosa.display
import soundfile as sf
import os
import matplotlib.pyplot as plt
import numpy as np

y, sr = librosa.load("../../test2.m4a")
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref = np.max)
img = librosa.display.specshow(D,y_axis = 'linear', x_axis = 'time', sr = sr)

plt.plot()

onsets = []

onsets = librosa.onset.onset_detect(y, backtrack = True, units = 'time')

print(onsets)
print(onsets.shape)
```

```python
i = 1
for onset in onsets:
    current_sample, sr = librosa.load("../../test2.m4a", offset = onset, duration = .3)
    sf.write(f'./rhythmcomps/noise{i}.wav', current_sample, sr)
    i+=1

print("complete")
```
